```python
def replace_all(text, codes):
    for i, j in codes.items():
        text = text.replace(i, j)
    return text
```

```python
codes = {
"EX": "Extinct",
"E": "Endangered",
"PE": "Extinct in the wild",
"CR": "Critically endangered",
"V": "Vulnerable",
"NT": "Near threatened wildlife",
"SL": "Special least concern",
"C": "Least concern wildlife",
"I": "International wildlife" 
}
text = data['status'] 
a = replace_all(text, codes)
data['status']  = a
```

```python
codes = {
"EX" : "Extinct",
"WX" : "Extinct in the wild", 
"CE" : "Critically endangered", 
"E" : "Endangered", 
"V" : "Vulnerable",
"CD" : "Conservation dependent"
}
text = data['EPBCstatus'] 
b = replace_all(text, codes)
data['EPBCstatus']  = b
```

```python
codes = {
"Q" : "Queensland endemic",
"QA" : "Australian endemic", 
"QI" : "Naturally occurs in Queensland and overseas", 
"QAI" : "Naturally occurs in Australia and overseas", 
"V" : "Native to Queensland",
"U" : "Native to Queensland - endemicity not defined",
"II" : "Naturalised from overseas",
"IA" : "Naturalised from elsewhere in Australia",
"IU" : "Naturalised species - origin not defined",
"VA" : "Australian vagrant",
"VI" : "International vagrant",
"VU" : "vagrant - origin not defined"
}
text = data['endemicity'] 
c = replace_all(text, codes)
data['endemicity']  = c
```

```python

codes = {
"Y" : "Yes",
"N" : "No"
}
text = data['confidential'] 
d = replace_all(text, codes)
data['confidential']  = d
```

```python

codes = {
"Y" : "Yes",
"N" : "No"
}
text = data['conservationSignificant'] 
d = replace_all(text, codes)
data['conservationSignificant']  = d
```
