```python
#import essential libraries 
import pandas as pd
import numpy as np
import dateutil.parser as parser
import json as json
```

```python
# requires openpyxl version 2.4.0+

from openpyxl import load_workbook
wb = load_workbook(filename='/Users/cha801/Documents/UpdateLists/EPBCActListofThreatened-Fauna.xlsx')
ws = wb['Sheet1'] # ws is now an IterableWorksheet


for cell in ws['A:A']:
    #print('Printing from ' + str(cell.column) + str(cell.row))
    ws.cell(row=cell.row, column=6, value=cell.value) 
    
# iterate thru all cells and if hyperlink found attempt modification of cell
for row in ws.rows:
    for cell in row:
        try:
            if len(cell.hyperlink.target)  > 0:
                cell.value = "".join([cell.hyperlink.target])
                # Join cell.value and hyperlink target into string (optionally just assign the hyperlink.target to the cell.value
        except:
            pass

# save workbook to temp .xlsx
wb.save("/Users/cha801/Documents/UpdateLists/v1/EPBCActListofThreatened-Fauna.xlsx")

# read with pandas 
data = pd.read_excel("/Users/cha801/Documents/UpdateLists/EPBCActListofThreatened-Fauna.xlsx")
```

```python
#rename the DwC terms
data = data.rename(columns=
{
'Genus, species (subspecies, population)':'scientificNameLink',
'Genus, species (subspecies, population).1':'scientificName',
'Common Name':'vernacularName',
'Status':'status',
'Effective':'effectiveDate',
'animal':'taxonRemark'
})
```

```python
#create copy for sourceStatus
data['sourceStatus'] = data['status'].copy()
```

```python
data['status'] = data['status'].str.split('(').str[0]
```

```python
data['effectiveDate'] = pd.to_datetime(data['effectiveDate'],dayfirst=True).dt.strftime("%Y-%m-%d")
```
